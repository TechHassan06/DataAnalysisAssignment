## QUESTION NUMBER 3

```python
df.MSRP
```

```python
avg_msrp = df.groupby('Make')['MSRP'].mean()

# Sort descending and take top 10
top_10_avg_msrp = avg_msrp.sort_values(ascending=False).head(10).round(0)
print(top_10_avg_msrp)
```

```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(10,6))
top_10_avg_msrp.plot(kind='barh', color='skyblue')
plt.xlabel('Average MSRP ($)')
plt.ylabel('Car Brand')
plt.title('Top 10 Average MSRP per Car Brand')
plt.gca().invert_yaxis() # To have highest at the top
plt.show()
```

## QUESTION NUMBER 4

```python
avg_msrp_per_year = df.groupby('Year')['MSRP'].mean().round(0)

# Print to verify
print(avg_msrp_per_year)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(avg_msrp_per_year.index, avg_msrp_per_year.values, marker='o', color='purple')
plt.xlabel('Year')
plt.ylabel('Average MSRP ($)')
plt.title('Average MSRP Over the Years')
plt.grid(True)
plt.show()
```

## QUESTION NUMBER 5

```python
df['Avg MPG'] = (df['highway MPG'] + df['city mpg']) / 2
# Group by Year and compute average Engine HP and average MPG
avg_stats = df.groupby('Year').agg({
    'Engine HP': 'mean',
    'Avg MPG': 'mean'
}).round(0)

print(avg_stats)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

# Plot Engine HP
plt.plot(avg_stats.index, avg_stats['Engine HP'], marker='o', color='red', label='Average Engine HP')

# Plot Average MPG
plt.plot(avg_stats.index, avg_stats['Avg MPG'], marker='s', color='green', label='Average MPG')

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('How Engine HP and MPG have changed over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

## QUESTION NUMBER 6

```python
print(df['Vehicle Size'].value_counts())
```

```python
# Count total number of cars
total_cars = len(df)

# Filter for Compact, Midsize, Large
size_counts = df['Vehicle Size'].value_counts()

# Get counts for desired categories
wanted_sizes = size_counts[['Compact', 'Midsize', 'Large']]

# Compute percentages
size_percentages = (wanted_sizes / total_cars) * 100
print(size_percentages.round(2))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.pie(size_percentages, labels=size_percentages.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('Percentage of Cars by Vehicle Size (Compact, Midsize, Large)')
plt.show()
```
