```python
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
```

```python
outliers_daily_usage = find_outliers_iqr(df, 'Daily_Usage_Hours')
outliers_exercise = find_outliers_iqr(df, 'Exercise_Hours')
outliers_screen_time = find_outliers_iqr(df, 'Screen_Time_Before_Bed')
```

# Question 4

```python
df = pd.read_csv("teen_phone_addiction_dataset.csv")

# Select only the relevant numeric columns
columns = [
    'Daily_Usage_Hours', 'Sleep_Hours', 'Academic_Performance', 'Addiction_Level',
    'Time_on_Social_Media', 'Time_on_Gaming', 'Time_on_Education', 'Apps_Used_Daily',
    'Anxiety_Level', 'Depression_Level', 'Self_Esteem', 'Social_Interactions',
    'Exercise_Hours', 'Screen_Time_Before_Bed', 'Phone_Checks_Per_Day',
    'Family_Communication', 'Weekend_Usage_Hours'
]

# Drop rows with missing values in these columns
data = df[columns].dropna()

# Compute correlation matrix
corr_matrix = data.corr()

# Plot correlation heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap of Teen Behaviors & Wellbeing')
plt.show()

# Flatten the correlation matrix to find pairs
corr_pairs = corr_matrix.unstack()

# Remove self correlations (correlation of a variable with itself)
corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]

# Drop duplicate pairs (since corr(A,B) == corr(B,A))
corr_pairs = corr_pairs.groupby(frozenset).first()

# Sort by correlation value
sorted_corr = corr_pairs.sort_values()

# Get top 2 negative and top 2 positive correlations
top_negatives = sorted_corr.head(2)
top_positives = sorted_corr.tail(2)

# Display results
print("Top 2 Strong Negative Correlations:")
print(top_negatives)

print("\nTop 2 Strong Positive Correlations:")
print(top_positives)
print("🔴 Top 2 Strong Negative Correlations:")
for (var1, var2), corr_value in top_negatives.items():
    print(f"- {var1} vs {var2}: {corr_value:.2f}")
    if "Addiction_Level" in [var1, var2] and "Sleep_Hours" in [var1, var2]:
        print("  ➔ Teens with higher addiction scores tend to sleep slightly less.")
    elif "Weekend_Usage_Hours" in [var1, var2] and "Self_Esteem" in [var1, var2]:
        print("  ➔ Slight indication that more weekend phone use might relate to lower self-esteem.")

print("\n🔵 Top 2 Strong Positive Correlations:")
for (var1, var2), corr_value in top_positives.items():
    print(f"- {var1} vs {var2}: {corr_value:.2f}")
    if "Daily_Usage_Hours" in [var1, var2] and "Addiction_Level" in [var1, var2]:
        print("  ➔ Teens who use their phones more hours daily tend to have higher addiction scores.")
    elif "Addiction_Level" in [var1, var2] and "Apps_Used_Daily" in [var1, var2]:
        print("  ➔ Teens with higher addiction also tend to use more apps daily.")

# Overall comment
print("\n Overall Interpretation:")
print("- More daily phone use and more apps used daily are clearly linked with higher addiction levels.")
print("- There is a mild tendency for higher addiction levels to reduce sleep hours.")
print("- Weekend usage shows a tiny link to lower self-esteem, but the relationship is very weak.")


```
